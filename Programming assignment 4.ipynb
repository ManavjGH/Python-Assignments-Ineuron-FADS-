#4
 A number is thought of as an Armstrong number if the sum of its own digits raised to the power number of digits gives the number itself.

```python
num=int(input('Enter a number: '))
sum=0
order=len(str(num))
copy_num=num
for i in range(1,num+1):
    digit=num%10
    sum=sum+digit**order
    num //= 10
    
print(sum)
if sum==copy_num:
    print(f'{copy_num} is a armstrong number')
else:
    print(f'{copy_num} is not a armstrong number')
```
